# Value at Risk Monte Carlo Model

## Importing Data

* Yahoo Finacial API
* Data stored in data frame with datetime index

```python
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd

# Create an empty portfolio dataframe
portfolio = pd.DataFrame()

# Set the start and end date
start_date = '2020-01-01'
end_date = '2022-01-01'

# Set the tickers for portfolio
tickers = ['GOOGL','AAPL','GME']

# Build portfolio
for ticker in tickers :
    
    # Get the data
    data = yf.download(ticker, start_date, end_date)

    # Add daily adjusted close prices portfolio
    portfolio[ticker] = data['Adj Close']

# Set time index
portfolio.set_index = data.index
    
# Print first five rows
print(portfolio.head())
```

## Display Stock Prices Over Time

```python
portfolio.plot(subplots = True, ylabel = 'Dollars ($)', title = 'Stock Prices');
```

## Display Stock Returns Over Time
$R_{t} = \frac{P_{t} - P_{t-1}}{P_{t-1}}$
* $R_{t}$: Returns at time $t$
* $P_{t-1}$: Price at time $t - 1$
* $P_{t}$: Price at time $t$
* $t$: Unit of time (here, days is used)

```python
portfolio.pct_change().mul(100).plot(subplots = True, ylabel = 'Returns (%)', title = 'Stock Returns');
```

## Create a Portfolio and Plot Loss
$L_{t} = -R_{t}$ 
* $L_{t}$: Loss at time $t$
* $R_{t}$: Return at time $t$

```python
# number of assests in portfolio
N = len(portfolio.columns)

# weights (all assests equally weights) 1 x N
weights = [1/N for w in range(N)]

# Plot each loss overtime
portfolio.pct_change().dropna().mul(-100).plot(subplots = True, title = 'Daily Losses',ylabel = 'Losses (%)')
plt.show()

# Plot histograms of each loss
portfolio.pct_change().dropna().mul(-1).plot(kind = 'hist', subplots = True, bins = 100, title = 'Daily Losses Distribution')
plt.show()
```

## Monte Carlo Simulation - Setup

```python
# Total steps for calculating daily losses
total_steps = 1440
steps = 1/total_steps

# Number of assets in portfolio
num_assets = len(portfolio.columns)

# Number of Monte Carlo simulations
N = 10000

# Create a daily loss matrix for all assets accross N runs
daily_loss = np.zeros((num_assets, N))

# Get the lower triangular matrix of the covariance matrix using the Cholesky method
cov = portfolio.pct_change().dropna().cov()
e_cov = np.linalg.cholesky(cov);

# Get average loss of all portfolio assests (num_assets, 1)
mu = np.expand_dims(portfolio.pct_change().dropna().mul(-1).mean().to_numpy(),axis = 1)

print(mu)
print(e_cov)
```

## Monte Carlo Simulation - Method 1: Affine Transformation 
$s = \mu + Az$
* $s$: Multivariate Normal Distribution Sample
* $\mu$: Vector of Means
* $A$: Lower Triangular Matrix of Covariance Matrix
* $z$: Vector of Normal Standard Distribution Samples

```python
from scipy.stats import norm
from scipy.stats import multivariate_normal
from pypfopt.risk_models import CovarianceShrinkage
import numpy as np

# Monte Carlo simulation with a Guassian distribution and N runs
for n in range(N):
    correlated_randomness = e_cov @ norm.rvs(size = (num_assets,total_steps))
    minute_losses = mu * steps + correlated_randomness * np.sqrt(steps)
    daily_loss[:,n] = minute_losses.sum(axis = 1)
    
losses = weights @ daily_loss
historical_losses = weights @ portfolio.pct_change().dropna().mul(-1).transpose().to_numpy()

# Plot
f, (ax1,ax2) = plt.subplots(2,1,sharex = True);
ax1.hist(losses,bins = 100);
ax2.hist(historical_losses, bins = 100);
ax1.title.set_text("Daily Losses - Simulated")
ax2.title.set_text("Daily Losses - Data")
ax2.set_xlabel("Fraction of Daily Loss")
ax1.set_ylabel("Count");
ax2.set_ylabel("Count");
```

## Monte Carlo Simulation - Method 2: Multivariate Normal Distribution
$s = \Phi(\Sigma,\mu)$
* $s$: Multivariate Normal Distribution Sample
* $\mu$: Vector of Means
* $\Sigma$: Covariance Matrix
* $\Phi$: Multivariate Normal Distribution

```python
mu = portfolio.pct_change().dropna().mul(-1).mean().to_numpy()

# Monte Carlo simulation with a Guassian distribution and N runs
for n in range(N):
    minute_losses = multivariate_normal.rvs(mean = mu * steps, cov = cov * steps, size = (total_steps)).transpose()
    daily_loss[:,n] = minute_losses.sum(axis = 1)

losses = weights @ daily_loss
historical_losses = weights @ portfolio.pct_change().dropna().mul(-1).transpose().to_numpy()

# Plot
f, (ax1,ax2) = plt.subplots(2,1,sharex = True);
ax1.hist(losses,bins = 100);
ax2.hist(historical_losses, bins = 100);
ax1.title.set_text("Daily Losses - Simulated")
ax2.title.set_text("Daily Losses - Data")
ax2.set_xlabel("Fraction of Daily Loss")
ax1.set_ylabel("Count");
ax2.set_ylabel("Count");
```

## Monte Carlo Simulation - Method 3: Multivariate Student's T Distribution

```python
from scipy.stats import multivariate_t
mu = portfolio.pct_change().dropna().mul(-1).mean().to_numpy()

# Monte Carlo simulation with a Guassian distribution and N runs
for n in range(N):
    minute_losses = multivariate_t.rvs(loc = mu * steps, shape = cov * steps, df = 4.0, size = (total_steps)).transpose()
    daily_loss[:,n] = minute_losses.sum(axis = 1)

losses = weights @ daily_loss
historical_losses = weights @ portfolio.pct_change().dropna().mul(-1).transpose().to_numpy()

# Plot
f, (ax1,ax2) = plt.subplots(2,1,sharex = True);
ax1.hist(losses,bins = 100, range = (-0.25,0.25));
ax2.hist(historical_losses, bins = 100, range = (-0.25,0.25));
ax1.title.set_text("Daily Losses - Simulated")
ax2.title.set_text("Daily Losses - Data")
ax2.set_xlabel("Fraction of Daily Loss")
ax1.set_ylabel("Count");
ax2.set_ylabel("Count");
```

## Monte Carlo Simulation - Simulate the Stock Market

```python
# Get prices from loss by assuming initial prices equal to one dollar
daily_prices = np.cumprod((1 - daily_loss), axis = 1)
print(daily_prices.shape)
plt.plot(np.transpose(daily_prices[0,:100]), label = 'GOOGL');
plt.plot(np.transpose(daily_prices[1,:100]), label = 'AAPL');
plt.plot(np.transpose(daily_prices[2,:100]), label = 'GME');
plt.xlabel('Day')
plt.ylabel('Stock Ammount ($)')
plt.legend()
plt.show();
```
